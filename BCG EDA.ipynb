```python
def annotate_stacked_bars(ax,pad=0.99,colour = "white",textsize = 13):
    for bar in ax.patches:
        height = round(bar.get_height(),1)
        if height == 0:
            continue
        ax.annotate(
            str(height),
            (bar.get_x()+bar.get_width()/2 , bar.get_y()+bar.get_height()/2),
            ha = 'center',
            va = 'center',
           
            size = textsize
        )
```

```python
ax = churn_percentage.transpose().plot(kind = 'bar',stacked = True,figsize = (8,6),rot = 0)
annotate_stacked_bars(ax,textsize = 14)
plt.legend(['Retention','Churn'],loc = 'upper right')
plt.ylabel('company base(%)')
```

```python
#nearly 10 % of the customers have churned which is an imbalance in the set
#we can check the sales channel for checking the main reason for the churn flaw in the sales chains 
#selecting relevant column 
channel = client_df[['id','channel_sales','churn']]
#grouping and unstacking 
#groups the data channel sales and churn, counts the no of 0 and 1 for a channel sales 0 is rentention and 1 is churn
channel = channel.groupby([channel['channel_sales'],channel['churn']])['id'].count().unstack(level = 1).fillna(0)
#counting total customer per channel if 
#one channel had 4 customer,then its the total,then figureout percentage of customer stayes and left 
#arrange them in order highest percentage to the left
#div(channel.sum(axis=1), axis=0): Divides each count by the total number of clients for that sales channel (row-wise sum).

channel_churn =  (channel.div(channel.sum(axis = 1),axis = 0)*100).sort_values(by=[1],ascending = False)
ax = channel_churn.plot(kind = 'bar',stacked = True,figsize=(8,6),rot =75)
annotate_stacked_bars(ax,textsize = 14)
plt.title('sales channel')
plt.legend(['retention','churn'],loc = 'upper right')
plt.ylabel('company base(%)')
```

```python
def plot_distribution(dataframe, column, ax, bins_=50):
    # Create a temporary DataFrame with the data to plot
    temp = pd.DataFrame({"Retention": dataframe[dataframe["churn"] == 0][column],
                         "Churn": dataframe[dataframe["churn"] == 1][column]})
    # Plot histogram
    temp[["Retention", "Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    # x-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')

# Now the rest of your code should work as expected
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])
```

```python
#plot historam 
#target_variable = the column name of the variable to plot in the histogram.
#if_churn: A boolean flag that determines whether to separate the histogram by the 'churn' column (default is True).
def plot_histogram_by_churn(df, target_variable, figsize=(20,5), bins=100, if_churn=True):
    fig = plt.figure(figsize=figsize)
    if if_churn:
        ax = sns.histplot(data=df, x=target_variable, bins=bins, hue='churn')
    else:
        ax = sns.histplot(data=df, x=target_variable, bins=bins)
```

```python
for attr in ['price_off_peak_var', 'price_peak_var', 'price_mid_peak_var', 'price_off_peak_fix', 'price_peak_fix', 'price_mid_peak_fix']:
    plot_histogram_by_churn(df=price_df, target_variable=attr, if_churn=False)
```

```python
#analyse price sensivity based on off peak price
#hypothesis investigation 
#checking the customer sensitivity to price when price increase some customers leave then we say these customers are sensitive to price as the increse in price lead the customer churn.i case there is decrease or no change in the price and customers still swtich it is hard to say customers on depend on the price.
#we have consumption data for each of the companies for the year 2014 we will create a new feature to measue price senssitivty using the average of the year the last 6 months and last 3 months 
```

```python
#create mean average data 
mean_year = price_df.groupby(['id']).mean().reset_index()
mean_6m = price_df[price_df['price_date'] > '2015-06-01'].groupby(['id']).mean().reset_index()
mean_3m = price_df[price_df['price_date'] > '2015-10-01'].groupby(['id']).mean().reset_index()
```

```python
mean_year.head()
```

```python
#rename the columns of mean year
mean_year = mean_year.rename(

    columns={
        "price_off_peak_var": "mean_year_price_off_peak_var",
        "price_peak_var": "mean_year_price_peak_var",
        "price_mid_peak_var": "mean_year_price_mid_peak_var",
        "price_off_peak_fix": "mean_year_price_off_peak_fix",
        "price_peak_fix": "mean_year_price_peak_fix",
        "price_mid_peak_fix": "mean_year_price_mid_peak_fix"
    }
)
```

```python
mean_year["mean_year_price_off_peak"] = mean_year["mean_year_price_off_peak_var"] + mean_year["mean_year_price_off_peak_fix"]
mean_year["mean_year_price_peak"] = mean_year["mean_year_price_peak_var"] + mean_year["mean_year_price_peak_fix"]
mean_year["mean_year_price_med_peak"] = mean_year["mean_year_price_mid_peak_var"] + mean_year["mean_year_price_mid_peak_fix"]
```

```python
#rename the columns of mean 6 month
mean_6m = mean_6m.rename(

    columns={
        "price_off_peak_var": "mean_6m_price_off_peak_var",
        "price_peak_var": "mean_6m_price_peak_var",
        "price_mid_peak_var": "mean_6m_price_mid_peak_var",
        "price_off_peak_fix": "mean_6m_price_off_peak_fix",
        "price_peak_fix": "mean_6m_price_peak_fix",
        "price_mid_peak_fix": "mean_6m_price_mid_peak_fix"
    }
)

mean_6m["mean_6m_price_off_peak"] = mean_6m["mean_6m_price_off_peak_var"] + mean_6m["mean_6m_price_off_peak_fix"]
mean_6m["mean_6m_price_peak"] = mean_6m["mean_6m_price_peak_var"] + mean_6m["mean_6m_price_peak_fix"]
mean_6m["mean_6m_price_med_peak"] = mean_6m["mean_6m_price_mid_peak_var"] + mean_6m["mean_6m_price_mid_peak_fix"]
```

```python
#rename the columns of mean 3 month
mean_3m = mean_3m.rename(

    columns={
        "price_off_peak_var": "mean_3m_price_off_peak_var",
        "price_peak_var": "mean_3m_price_peak_var",
        "price_mid_peak_var": "mean_3m_price_mid_peak_var",
        "price_off_peak_fix": "mean_3m_price_off_peak_fix",
        "price_peak_fix": "mean_3m_price_peak_fix",
        "price_mid_peak_fix": "mean_3m_price_mid_peak_fix"
    }
)

mean_3m["mean_3m_price_off_peak"] = mean_3m["mean_3m_price_off_peak_var"] + mean_3m["mean_3m_price_off_peak_fix"]
mean_3m["mean_3m_price_peak"] = mean_3m["mean_3m_price_peak_var"] + mean_3m["mean_3m_price_peak_fix"]
mean_3m["mean_3m_price_med_peak"] = mean_3m["mean_3m_price_mid_peak_var"] + mean_3m["mean_3m_price_mid_peak_fix"]
```

```python
#merge into 1 dataframe 
price_features = pd.merge(mean_year,mean_6m, on = 'id')
price_features = pd.merge(price_features,mean_3m, on = 'id')
```

```python
price_features.head()
```

```python
#merge the price feature dataset with churn variable to check wheather price sensitivity has any correlation with churn 
#he merging of the price features dataset with the churn variable is performed to analyze whether there is a correlation between price sensitivity and customer churn. Here's a detailed explanation of why this is done:

#Understanding Customer Behavior:

##This helps in understanding if price sensitivity is a significant factor in customer retention.
#Data Analysis and Insights:

#Merging allows you to perform statistical analysis, such as calculating correlation coefficients or conducting hypothesis tests, to see if there is a relationship between price features and churn.
#Insights gained can be used to adjust pricing strategies to minimize churn.
#Feature Engineering for Predictive Modeling:

#If you're building a predictive model to predict churn, having price features alongside the churn variable can improve the model's accuracy.
#This enriched dataset provides more variables for the model to learn from, potentially leading to better predictions.
#Customer Segmentation:

#Analyzing this merged data can help segment customers based on their price sensitivity and churn risk.
#This segmentation can inform targeted marketing and retention strategies, such as personalized offers or discounts for price-sensitive customers.
```

```python
price_churn = pd.merge(price_features,client_df[['id','churn']],on = 'id')
price_churn.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'price_churn' is your DataFrame

# Select only the numeric columns
numeric_df = price_churn.select_dtypes(include=['number'])

# Calculate correlation matrix
corr = numeric_df.corr()

# Plot correlation heatmap
plt.figure(figsize=(20, 18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, cmap='Greens', annot=True, annot_kws={'size': 10})

# Axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

```python
#correlation between price variable with each other and with churn,correlation betchurn and prices variable is very low 
#which means customers churn is not sensitive to price change 
```
